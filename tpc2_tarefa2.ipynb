### <font color = 'green'>World Cup with Pandas</font>

```python
# Read the Players and Teams data into dataframes
Op = open('Players.csv')
playersDF = pd.read_csv(Op)
Op = open('Teams.csv')
teamsDF = pd.read_csv(Op)
```

```python
#Qual jogador (print do ultimo nome) de uma equipa com "ia" no nome, 
#que jogou menos de 200 minutos e fez mais de 100 passes?
Jogador = playersDF[playersDF.team.str.contains("ia") & (playersDF.minutes < 200) & (playersDF.passes > 100)]
Jogador
```

```python
# Qual é o número médio de passes feitos pelos avançados? Pelos meio-campistas?
# Verifique se a resposta especifica qual é qual e não inclua outras
# posições no seu resultado.


passesmedios = playersDF.loc[playersDF["position"]=="midfielder"]["passes"].mean()
passesavançados = playersDF.loc[playersDF["position"]=="forward"]["passes"].mean()
    
    
print(((str(passesmedios)) + " é numero médio de passes feitos por jogadores médios"))
print(((str(passesavançados)) + " é numero média de passes feitos por jogadores avançados"))

```

```python
# Qual equipe tem a maior relação (razão, racio, etc) de goalsFor versus goalsAgainst?
# Imprima apenas o nome da equipa.
# Dica: adicione uma coluna "racio" ao dataframe das equipas e, depois, faça o seu "sort" e, 
# em seguida, use a head(1) ou tail(1) ...

teamsDF["Racio"] = teamsDF.goalsFor/teamsDF.goalsAgainst
teamsDFsorted = teamsDF.sort_values(\
                        by = "Racio",\
                        inplace = False,
                        ascending = False)
teamsDFsorted.reset_index(drop = False, inplace = True)

teamsDFsorted.team.iloc[0]
```

```python
# Quantos jogadores que jogam numa equipa com ranking <10, 
# jogaram mais de 350 minutos?
# Lembrete: len() fornece o número de linhas que um dataframe contém


#defenir uma variavel teams com valor de equip com rank menos 10
teams = teamsDF.loc[teamsDF["ranking"]<10]["team"].values

# buscar jogadores que estão na variavel defenida anteriormente
jogador = playersDF.iloc[[team for team in range(len(playersDF["team"])) if playersDF["team"][team] in teams]]

# defenir variavel que contem apenas quem esta acima dos 350 minutos
jogador = jogador.loc[jogador["minutes"]>350]

# print de da string com comprimento do numero integral de jogadores
print( str(len(jogador)) + " jogadores fizeram mais de 350 minutos")
jogador
```

### <font color = 'green'>Titanic with Pandas</font>

```python
# Read the Titanic data into a dataframe
f = open('Titanic.csv')
titanicDF = pd.read_csv(f)
titanicDF
```

```python
# Quantas mulheres casadas embarcaram em Cherbourg?
# Nota: 'first' é uma função no Pandas, então 'titanic.first' irá gerar um erro;
# use 'titanic[' first ']

mulher_casada = titanicDF.iloc[[value for value in range(len(titanicDF["first"].values))
                                if "Mrs" in titanicDF["first"].values[value]]]

embarq_cherbourg = mulher_casada.loc[mulher_casada["embarked"]=="Cherbourg"]

print("As mulheres casadas que embarcaram em Cherbourg foram: " + str(len(embarq_cherbourg)))


#Para perceber o numero de mulheres casadas, defenimos duas variáveis:
#Determinar as mulheres casadas através de uma variável,que se encontra no dataframe titanic, e na coluna "first" e perceber
#quais são as mulheres que apresentam "mrs" no seu nome, sendo por isso casadas. Se for verdade fica na variável.
#Na outra parte, defenimos que mulheres casadas embarcaram em Cherbourg.
#Para isto, definimos a variável embarq_cherbourg, e através da função loc, procuramos as mulhers que
#dentro da variável anterior tinham o valor "cherbourg" no "embarked". E através do print, conseguimos ver
#a quantidade de mulheres casadas que embarcaram em "cherbourg".
```
