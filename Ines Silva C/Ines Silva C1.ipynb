```python
#Importar libraries
import pandas as pd
import os
import matplotlib.pyplot as plt

#Carregar os dados do ficheiro .csv
df_exel = pd.read_csv("company_sales_data.csv");

#Criar as listas
profitList = df_exel['total_profit'].tolist() ;
monthList = df_exel['month_number'].tolist();
```

```python
#Exercicio 1

data = plt.axes();

data.set(xlim=(1,12),ylim=(100000,500000),
        xlabel='Month number',ylabel='Profit in dollar',
        title='Company profit per month');

data.plot(monthList, profitList);
```

3- Mostrar as vendas mensais de todos os produtos num gráfico de múltiplas linhas, ou seja, uma linha diferente para cada produto.

```python
#Aqui vamos buscar os dados
lista_shampoo = df_exel['shampoo'].tolist();
lista_moisturizer = df_exel['moisturizer'].tolist();
lista_facecream = df_exel['facecream'].tolist();
lista_facewash = df_exel['facewash'].tolist();
lista_toothpaste = df_exel['toothpaste'].tolist();
lista_bathingsoap = df_exel['bathingsoap'].tolist();

#Referente ao exercicio 3

data3 = plt.axes();
data3.set(xlim=(1,12),ylim=(1000,18000),
        xlabel='Sales units in number',ylabel='Month Number',
        title='Sales Data');

#Linhas gráfico

toothpastelinha, = data3.plot(monthList, lista_toothpaste, color='green', marker='o', markerfacecolor='green',label='Thooth Paste Sales Data');
bathingsoaplinha, = data3.plot(monthList, lista_bathingsoap, color='red', marker='o', markerfacecolor='red',label='Bathing Soap Sales Data');
shampoolinha, = data3.plot(monthList, lista_shampoo, color='purple', marker='o', markerfacecolor='purple',label='Shampoo Sales Data');
facecreamlinha, = data3.plot(monthList, lista_facecream, color='blue', marker='o', markerfacecolor='blue',label='Face cream Sales Data');
facewashlinha, = data3.plot(monthList, lista_facewash, color='orange', marker='v', markerfacecolor='orange',label='Face wash Sales Data');
moisturizerlinha, = data3.plot(monthList, lista_moisturizer, color='brown', marker='o', markerfacecolor='brown',label='Moisturizer Sales Data');


#legenda do Gráfico

data3.legend(handles=[facecreamlinha, facewashlinha, toothpastelinha, bathingsoaplinha, shampoolinha, moisturizerlinha],loc="upper left");

```

4- Calcular para cada produto o total das vendas do ano e mostre-os
usando Pie Chart (gráfico de queijo).
No Pie chart inscrever a % de vendas de cada produto.

```python
#Exercico 4
t_facewash = sum(lista_facewash)
t_facecream = sum(lista_facecream)
t_shampoo = sum(lista_shampoo)
t_moisturizer = sum(lista_moisturizer)
t_toothpaste = sum(lista_toothpaste)
t_bathingsoap = sum(lista_bathingsoap)

t_produtos = t_facecream + t_bathingsoap + t_shampoo + t_facewash + t_toothpaste + t_moisturizer

facewash_percent = (t_facewash * 100)/t_produtos
facecream_percent = (t_facecream * 100)/t_produtos
shampoo_percent = (t_shampoo * 100)/t_produtos
moisturizer_percent = (t_moisturizer * 100)/t_produtos
toothpaste_percent = (t_toothpaste * 100)/t_produtos
bathingsoap_percent = (t_bathingsoap * 100)/t_produtos

# Data para plot

sizes = [facecream_percent, facewash_percent, toothpaste_percent, bathingsoap_percent, shampoo_percent, moisturizer_percent]
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown']
labels = 'Face Cream', 'Face Wash', 'Tooth Paste', 'Bathing Soap', 'Shampoo', 'Moisturizer'
explode = (0, 0, 0, 0, 0, 0) 

#Caraterizar
patches, texts = plt.pie( sizes, colors = colors, shadow = True, startangle = 90 )
#definir as diferentes etiquetas para as divisões do grafico circular
labels = ['Face Cream', 'Face Wash', 'Tooth Paste', 'Bathing Soap', 'Shampoo', 'Moisturizer']
plt.legend(patches, labels, loc="lower right")


#Fazer Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,

autopct='%1.1f%%', shadow=False, startangle=0)
#impor limites no gráfico,
plt.axis('equal')
plt.show()
```
